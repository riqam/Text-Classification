```python
import joblib

def load_data(return_file=False):
    X_train = joblib.load('Test Kompas\X_train.pkl')
    y_train = joblib.load('Test Kompas\y_train.pkl')
    X_valid = joblib.load('Test Kompas\X_valid.pkl')
    y_valid = joblib.load('Test Kompas\y_train.pkl')
    X_test = joblib.load('Test Kompas\X_test.pkl')
    y_test = joblib.load('Test Kompas\y_test.pkl')

    if return_file:
        return X_train, y_train, X_valid, y_valid, X_test, y_test
```

```python
X_train, y_train, X_valid, y_valid, X_test, y_test = load_data(return_file=True)
```

```python
# Hapus kolom yang tidak diperlukan
columns_to_drop = ['Url', 'UrlShort', 'SiteID', 'SectionID', 'AuthorID', 'Photo', 'Lipsus', 'Video', 'EmbedSocial', 'Related', 'PublishedDate']
def drop_columns(return_file=False):
    X_train.drop(columns=columns_to_drop, inplace=True)
    X_valid.drop(columns=columns_to_drop, inplace=True)
    X_test.drop(columns=columns_to_drop, inplace=True)
    
    if return_file:
        return X_train, X_valid, X_test
    
drop_columns(return_file=True)
```

```python

def list_to_string(text):
    if isinstance(text, list):
        return ' '.join(text)
    else:
        return text
```

```python
X_train['Tag'] = X_train['Tag'].apply(list_to_string)
X_valid['Tag'] = X_valid['Tag'].apply(list_to_string)
X_test['Tag'] = X_test['Tag'].apply(list_to_string)
```

```python
import joblib
import re
import html as ihtml
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from bs4 import BeautifulSoup

from nltk.stem import PorterStemmer

#factory = StemmerFactory()
#stemmer = factory.create_stemmer()
stemmer = PorterStemmer()
stop_words = set(stopwords.words('indonesian'))  # Stopwords dalam bahasa Indonesia

def preprocess_text(text):
    # Memastikan bahwa nilai bukan None
    if text is not None:
        # Menghapus HTML tags
        text = BeautifulSoup(ihtml.unescape(text)).text
        text = re.sub(r"http[s]?://\S+", "", text)
        text = re.sub(r"\s+", " ", text) 

        # Lowercasing
        text = text.lower()

        # Menghapus karakter khusus
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text) # ^a-zA-Z\s

        # Tokenisasi
        tokens = word_tokenize(text)

        # Menghapus stop words
        tokens = [word for word in tokens if word not in stop_words]

        # Stemming
        tokens = [stemmer.stem(word) for word in tokens]

        # Menggabungkan token kembali menjadi teks
        preprocessed_text = ' '.join(tokens)

        return preprocessed_text
    else:
        return ''  # Mengembalikan string kosong untuk nilai None
```

```python
X_train = X_train.applymap(preprocess_text)
X_valid = X_valid.applymap(preprocess_text)
X_test = X_test.applymap(preprocess_text)

joblib.dump(X_train, 'Test Kompas\X_train_pre.pkl')
joblib.dump(X_valid, 'Test Kompas\X_valid_pre.pkl')
joblib.dump(X_test, 'Test Kompas\X_test_pre.pkl')
```

```python
X_train = joblib.load('Test Kompas\X_train_pre.pkl')
X_valid = joblib.load('Test Kompas\X_valid_pre.pkl')
X_test = joblib.load('Test Kompas\X_test_pre.pkl')
```

```python
def tokenization():
    X_train['tokens'] = X_train['FullText'].apply(lambda x: word_tokenize(x))
    X_valid['tokens'] = X_valid['FullText'].apply(lambda x: word_tokenize(x))
    X_test['tokens'] = X_test['FullText'].apply(lambda x: word_tokenize(x))

    return X_train, X_valid, X_test
```

```python
X_train, X_valid, X_test = tokenization()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def vectorizer():
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_vectorizer.fit(X_train['tokens'].apply(lambda x: ' '.join(x)))

    joblib.dump(tfidf_vectorizer, 'Test Kompas/tfidf_vectorizer.pkl')

    return tfidf_vectorizer
```

```python
vectorizer()
```

```python
def applying():
    X_text_train = joblib.load('tfidf_vectorizer.pkl').transform(X_train['tokens'].apply(lambda x: ' '.join(x)))
    X_text_valid = joblib.load('tfidf_vectorizer.pkl').transform(X_valid['tokens'].apply(lambda x: ' '.join(x)))
    X_text_test = joblib.load('tfidf_vectorizer.pkl').transform(X_test['tokens'].apply(lambda x: ' '.join(x)))

    # Concatenating sparse matrices
    X_train_vect = joblib.dump(X_text_train, 'X_train_vect.pkl')
    X_valid_vect = joblib.dump(X_text_valid, 'X_valid_vect.pkl')
    X_test_vect = joblib.dump(X_text_test, 'X_test_vect.pkl')

    return X_train_vect, X_valid_vect, X_test_vect
    
```

```python
X_train_vect, X_valid_vect, X_test_vect = applying()
```

```python
X_train_vect = joblib.load('X_train_vect.pkl')
X_valid_vect = joblib.load('X_valid_vect.pkl')
X_test_vect = joblib.load('X_test_vect.pkl')
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

y_train = joblib.load('Test Kompas\y_train.pkl')
y_valid = joblib.load('Test Kompas\y_valid.pkl')

clf = MultinomialNB()

# 6. Pelatihan model
clf.fit(X_train_vect, y_train)

# 7. Evaluasi model
y_pred = clf.predict(X_valid_vect)

print(classification_report(y_valid, y_pred))
```
